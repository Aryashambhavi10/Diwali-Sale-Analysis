# Gender

```python
df.columns
```

```python
ad= sns.countplot(x="Gender",data=df,width =0.2,hue="Gender")
for i in ad.containers:
    ad.bar_label(i)
```

```python
sales_data=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
py=sns.barplot(x=("Gender"),y="Amount",data=sales_data,width=0.2,hue="Gender")

for a in py.containers:
    py.bar_label(a)

```

# State

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(25,6)})
cv=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
bv=sns.barplot(x="State",y="Orders",width=0.2,data=cv,hue="State",errorbar=('ci',False))
for f in bv.containers:
    bv.bar_label(f)
```

```python
sns.set(rc={'figure.figsize':(25,5)})
cv=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
bv=sns.barplot(x="State",y="Amount",width=0.2,data=cv,hue="State",errorbar=('ci',False))
for x in bv.containers:
    bv.bar_label(x)


```
